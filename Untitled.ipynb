```python
df3 = pd.read_csv("data capstone/FAOSTAT_data_2-3-2024 (2) (1).csv")
df3.drop(0, inplace=True)
df3= pd.melt(df3,id_vars= ["Unnamed: 0","Unnamed: 1","Unnamed: 2"], var_name = "year" )
df3.dropna(inplace = True)
df3["country"] = df3["Unnamed: 0"]
df3["pesticide use per area of cropland"] = df3["value"]

df3.drop(["Unnamed: 0","Unnamed: 1","Unnamed: 2","value"], axis = 1, inplace = True)
df3.to_csv("Pesticide use.csv")
print(df3.shape)
print(df3.head())
```

```python
df4 = pd.read_csv("data capstone/FAOSTAT_data_2-3-2024 (3) (3).csv")
df4.drop(0, inplace=True)
df4= pd.melt(df4,id_vars= ["Unnamed: 0","Unnamed: 1","Unnamed: 2"], var_name = "year" )
df4.dropna(inplace = True)
def g(df):
    df["Import Quantity"] = df["Unnamed: 1"].str.contains("Quantity").astype(int) * df["value"]
    
    return df

df4 = g(df4)
df4.head()
```

```python
def ga(df):
    df["Import Value"] = df["Unnamed: 1"].str.contains("Value").astype(int) * df["value"]
    df.drop(["Unnamed: 1","value"], axis = 1, inplace = True)
    return df
df4 = ga(df4)
df4 = df4.groupby(['Unnamed: 0', 'year'], as_index=False).sum()
df4["country"] = df4["Unnamed: 0"]

df4.drop(["Unnamed: 0","Unnamed: 2"], axis = 1, inplace = True)
df4.to_csv("Import.csv")

print(df4.shape)
print(df4.head())

```

```python
df5=pd.read_csv("data capstone/FAOSTAT_data_2-3-2024 (4).csv")
df5.drop(0, inplace=True)
df5= pd.melt(df5,id_vars= ["Unnamed: 0","Unnamed: 1","Unnamed: 2"], var_name = "year" )
df5.dropna(inplace= True)

def h(df):
    df["Area harvested"] = df["Unnamed: 1"].str.contains("Area harvested").astype(int) * df["value"]
    df["Production"] = df["Unnamed: 1"].str.contains("Production").astype(int) * df["value"]
    df["Yield"] = df["Unnamed: 1"].str.contains("Yield").astype(int) * df["value"]
    df.drop("Unnamed: 1", axis=1, inplace=True)
    
    return df
df5 = h(df5)
df5= df5.groupby(['Unnamed: 0', 'year'], as_index=False).sum()
df5["country"] = df5["Unnamed: 0"]

df5.drop(["Unnamed: 0","Unnamed: 2","value"], axis = 1, inplace = True)
df5.to_csv("Yield.csv")
print(df5.shape)
df5.head()

```

```python
df6=pd.read_csv("data capstone/FAOSTAT_data_2-3-2024 (5).csv")
df6.drop(0, inplace=True)
df6= pd.melt(df6,id_vars= ["Unnamed: 0","Unnamed: 1","Unnamed: 2","Unnamed: 3"], var_name = "month",value_name="cpi" )

df6.dropna(inplace= True)
df6["country"] = df6["Unnamed: 0"]
df6["year"] = df6["Unnamed: 3"]
df6.drop(["Unnamed: 0","Unnamed: 1","Unnamed: 2","Unnamed: 3"], axis = 1, inplace = True)
df6["cpi"]= df6["cpi"].astype("float")
df6 = df6[["country", "year", "month", "cpi"]]
df6= df6.groupby(["country", "year"])["cpi"].mean().reset_index()
df6.to_csv("CPI.csv")

print(df6.shape)
df6.head(25)
```

```python
df7= pd.read_csv("data capstone/FAOSTAT_data_2-3-2024.csv")
df7.drop(0, inplace=True)
df7= df7[df7["Unnamed: 2"] == "Maize (corn)"]
df7= pd.melt(df7,id_vars= ["Unnamed: 0","Unnamed: 1","Unnamed: 2"], var_name = "year" )
df7.dropna(inplace= True)
df7["country"] = df7["Unnamed: 0"]
df7["maize GPI"] = df7["value"]

df7.drop(["Unnamed: 0","Unnamed: 1","Unnamed: 2","value"], axis = 1, inplace = True)
df7.to_csv("GPI.csv")

print(df7.shape)
df7.head(8)
```

```python
df8=pd.read_csv("data capstone/FAOSTAT_data_en_2-3-2024.csv")
df8.dropna(inplace = True)
def j(df):
    df["Cropland nitrogen per unit area"] = df["Element"].str.contains("Cropland nitrogen per unit area").astype(int) * df["Value"]
    df["Cropland phosphorous per unit area"] = df["Element"].str.contains("Cropland phosphorous per unit area").astype(int) * df["Value"]
    df["Cropland potassium per unit area"] = df["Element"].str.contains("Cropland potassium per unit area").astype(int) * df["Value"]
    df.drop("Element", axis=1, inplace=True)
    
    return df
df8 = j(df8)
df8= df8.groupby(['Area', 'Year'], as_index=False).sum()
df8["country"] = df8["Area"]

df8.drop(["Area","Value","Cropland phosphorous per unit area"], axis = 1, inplace = True)
df8.to_csv("Nutrients use.csv")
print(df8.shape)
df8.tail()
```
